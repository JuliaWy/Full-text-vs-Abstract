Ein Jupyter Notebook erstellen und die zwei Datensätze (train & test) via Pandas einlesen, Datensätze zusammenführen, 200 zufällige Papers auswählen.

```python
import pandas as pd
import numpy as np
```

```python
# Datensätze importieren
test = pd.read_csv("C:/Users/Julia/Desktop/Masterarbeit/Full Dataset/metastatic_local_test.csv")
train = pd.read_csv("C:/Users/Julia/Desktop/Masterarbeit/Full Dataset/metastatic_local_train.csv")
```

```python
#Datensätze zusammenführen
full = pd.concat([test, train])
```

```python
#Alle ASCO-Papers rausfiltern (alle die "JCO" in der doi haben) 
# Das ~ macht die verneninung
filtermask = ~full["doi"].str.contains("JCO")
full_noJCO = full[filtermask]
full_noJCO.head()
full_noJCO.tail()
```

```python
#Zufallszahl generieren
x = np.random.seed(86)
#200 Einträge zufällig auswählen
random200_noJCO = full_noJCO.sample(n = 200, random_state = x)
```

```python
#Anfang der Tabelle zeigen (schauen ob das mit dem immer gleichen Zufall klappt (250, 236))
random200_noJCO.head()
```

```python
#Excelfile (.csv Format, nicht.xlsx) erstellen mit den 200 zufälligen Papers
random200_noJCO.to_csv("C:/Users/Julia/Desktop/Masterarbeit/Full Dataset/metastatic_local_random200_noJCO.csv")
```

Read the two datasets from "Metastatic Versus Localized Disease" (test & train) via Pandas, merge datasets, select 200 random papers

```python
import pandas as pd
import numpy as np
```

```python
# Import datasets
test = pd.read_csv("C:/Users/Julia/Desktop/Full text vs. Abstract/1) dataset/metastatic_local_test.csv")
train = pd.read_csv("C:/Users/Julia/Desktop/Full text vs. Abstract/1) dataset/metastatic_local_train.csv")
```

```python
# Merge datasets
full = pd.concat([test, train])
```

```python
# Filter out all ASCO papers (all that have "JCO" in the DOI)
filtermask = ~full["doi"].str.contains("JCO") # The ~ operator performs negation
full_noJCO = full[filtermask]
full_noJCO.head()
```

```python
# Generate random seed
x = np.random.seed(86)

# Randomly select 200 entries
random200_noJCO = full_noJCO.sample(n = 200, random_state = x)
```

```python
# Show beginning of the table (check whether the fixed random seed works)
random200_noJCO.head()
```

```python
# Create Excel file (.csv format, not .xlsx) with the 200 random papers
random200_noJCO.to_csv("C:/Users/Julia/Desktop/Full text vs. Abstract/1) dataset/metastatic_local_random200_noJCO.csv")
```

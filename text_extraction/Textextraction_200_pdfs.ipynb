Text mit PyMuPdf aus den pdfs extrahieren und als zusätzliche Spalte an die CSV-Datei anhängen.

```python
import pandas as pd
from pathlib import Path
import pymupdf
```

```python
# Datensatz importieren
tabelle = pd.read_csv("C:/Users/Julia/Desktop/Masterarbeit/Textextraktion/metastatic_local_random200_noJCO.csv")
```

```python
#Neue Spalte (leer) anfügen
tabelle["full_text"] = None
```

```python
pdf_ordner = Path("C:/Users/Julia/Desktop/Masterarbeit/pdfs der Papers")
index = 0
for pdf_path in pdf_ordner.glob("*.pdf"):
    fulltext = "" #macht einen leeren String, wo dann der Text reinkommt
    pdf = pymupdf.open(pdf_path) #pdf öffnen
    for page in pdf: #Iteriert durch die pdf Seiten
        text = page.get_text() #Text der Seite extrahieren
        fulltext += text #hängt den String hinten an
    pdf.close()
    tabelle.loc[index, "full_text"] = fulltext
    index += 1
    
                                                                              
```

```python
#Nur Teil der Tabelle anzeigen lassen "subsetting"
tabelle[["doi", "accept", "full_text"]] #man macht eine liste in [[]] von den Spaltennamen, der Spalten die man will
```

```python
#Excelfile (.csv Format, nicht.xlsx) erstellen mitzusätzlicher Spaltee
tabelle.to_csv("C:/Users/Julia/Desktop/Masterarbeit/Textextraktion/metastatic_local_200_fulltext.csv")
```

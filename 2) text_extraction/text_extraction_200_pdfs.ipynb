Extract the full text from the PDFs using PyMuPDF and append it as an additional column to the CSV file


```python
import pandas as pd
from pathlib import Path
import pymupdf
```

```python
# Import dataset
table = pd.read_csv("C:/Users/Julia/Desktop/Full text vs. Abstract/2) text_extraction/metastatic_local_random200_noJCO.csv")
```

```python
#Add new (empty) column
table["full_text"] = None
```

```python
pdf_folder = Path("C:/Users/Julia/Desktop/Masterarbeit/pdfs der Papers")
index = 0
for pdf_path in pdf_folder.glob("*.pdf"):
    fulltext = "" # creates an empty string where the text will be stored
    pdf = pymupdf.open(pdf_path) # open PDF
    for page in pdf: # iterate through the PDF pages
        text = page.get_text() # extract page text
        fulltext += text # append text to string
    pdf.close()
    table.loc[index, "full_text"] = fulltext
    index += 1
    
                                                                              
```

```python
# Display only a subset of the table
table[["doi", "accept", "full_text"]]
```

```python
# Create CSV file (not .xlsx) with additional column with the full texts
table.to_csv("C:/Users/Julia/Desktop/Full text vs. Abstract/2) text_extraction/metastatic_local_200_fulltext.csv")
```

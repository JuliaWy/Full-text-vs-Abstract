```python
import os
from openai import OpenAI
import pandas as pd
```

```python
#mit Key auf ChatGPT zugreifen
os.environ["TestKey"] = Testkey #Statt TestKey den echten Key einsetzen!
client = OpenAI(api_key=os.getenv("TestKey"))
```

```python
#dataframe einlesen 
tabelle = pd.read_csv("C:/Users/Julia/Desktop/Masterarbeit/Textextraktion/metastatic_local_200_fulltext.csv") 
```

```python
#Funktion definieren
def classify(text_spalte):
    tabelle[text_spalte + "_gpt-5-2025-08"] = None
    for index, text in tabelle[text_spalte].items():
        response = client.chat.completions.create(
            model="gpt-5-2025-08-07", #Modellversion mit timestamp
            messages= [
                {"role": "system", "content": ( #Modell anpassen
                        "You will be provided with text excerpts from a cancer clinical trial."
                        "Your task is to classify patient eligibility based on disease stage. "
                        "Specifically, determine whether the trial included:\n"
                            "- Patients with localized or locally advanced disease, i.e. cancer that has not formed distant metastases\n"
                            "- Patients with metastatic disease\n"
                        "Your output must be a valid Python-list, this list may contain 0, 1 or 2 elements."
                        "The allowed elements are:\n"
                            "- \'LOCAL\' = Patients with localized or locally advanced disease were eligible for the trial"
                            "- \'METASTATIC\' = Patients with metastatic disease were eligible for the trial"
                        "If you are unsure, please provide your best guess."
                        "Do NOT return any additional text, exlpanations or punctuation marks outside the list."
                )},
                {"role": "user", "content": text} #"Prompt", da wird der Studientext geschickt
            ]
        )
        tabelle.loc[index, text_spalte + "_gpt-5-2025-08"] = response.choices[0].message.content #1. Anwort vom Modell wird in die Tabelle geschrieben
        print(index, response.choices[0].message.content)
```

```python
classify("abstract")
classify("full_text")
tabelle.head()
```
